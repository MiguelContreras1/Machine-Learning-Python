## EJERCICIO REGRESIÓN LINEAL MÚLTIPLE

```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generar datos random de x y y
rng = np.random.RandomState(1)
X = 10 * rng.rand(100, 3)
y = 0.5 + np.dot(X, [1.5, -2., 1.])
```

```python
X
```

```python
y
```

```python
# Instanciar modelo
model = LinearRegression(fit_intercept=True)
model
```

```python
# Entrenar el módelo
model.fit(X, y)
```

```python
# Ver el coeficiente de la pendiente (de las x)
print("Model slope: ", model.coef_)
```

```python
# Ver el coeficiente del intercepto
print("Model intercept:", model.intercept_)
```

```python
# Predicción para datos nuevos
xfit = np.linspace(-1, 11)
```

```python
Xfit = xfit[:, np.newaxis]
Xfit = np.hstack([Xfit, np.zeros((Xfit.shape[0], 2))])
```

```python
yfit = model.predict(Xfit)
```
